<h3>The Dot Product</h3>

<p>If we have two vectors (say, *x* and *y) with the same length n*, we can calculate the dot product between them. This is annotated several different ways, including the following:</p>
<ul>
    <li>xy</li>
    <li>$x^Ty$</li>
    <li>(left angle bracket)x,y(right angle bracket)</li>
</ul>

<p>Regardless which notation you use (I prefer the first), the calculation is the same; we calculate products in an element-wise fashion and then sum reductively across the products to a scalar value. That is: $$xy=\sum_{i=1}^nx_iy_i$$

The dot product is ubiquitous in deep learning: It is performed at every artificial neuron in a deep neural network, which may be made up of millions (or orders of magnitude more) of these neurons.</p>

<h3>Solving Linear Systems</h3>
<p>In the <b>Substitution</b> example, the two equations in the system are:</p>
<p>$$y = 3x$$</p>
<p>$$-5x + 2y = 2$$</p>
<p>The second equation can be rearranged to isolate <b>y</b>:</p>
<p>$$2y = 2 + 5x$$</p>
<p>$$y=\frac{2+5x}{2}=(1+\frac{5x}{2})$$</p>

```python
x = np.linspace(-10, 10, 1000)
```

```python
y1 = 3 * x
```

```python
y2 = 1 + (5 * x) / 2
```

```python
fig, ax = plt.subplots()
plt.xlabel('x')
plt.ylabel('y')
ax.set_xlim([0, 3])
ax.set_ylim([0, 8])
ax.plot(x, y1, c='green')
ax.plot(x, y2, c='brown')
plt.axvline(x=2, color='purple', linestyle='--')
_ = plt.axhline(y=6, color='purple', linestyle='--')
```

<p>In the <b>Elimination</b> example, the two equations in the system are:</p>
<p>$$2x-3y=15$$</p>
<p>$$4x+10y=14$$</p>
<p>Both equations can be rearranged to isolate <b>y</b>. Starting with the first equation:</p>
<p>$$-3y=15-2x$$</p>
<p>$$y=\frac{15-2x}{-3}=-5+\frac{2x}{3}$$</p>
<p>Then for the second equation:</p>
<p>$$4x+10y=14$$</p>
<p>$$2x+5y=7$$</p>
<p>$$5y=7-2x$$</p>
<p>$$y=\frac{7-2x}{5}$$</p>

```python
y1 = -5 + (2*x) / 3
```

```python
y2 = (7 - 2*x) / 5
```

```python
fig, ax = plt.subplots()
plt.xlabel('x')
plt.ylabel('y')

# Add x and y axes: 
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')

ax.set_xlim([-2, 10])
ax.set_ylim([-6, 4])
ax.plot(x, y1, c='green')
ax.plot(x, y2, c='brown')
plt.axvline(x=6, color='purple', linestyle='--')
_ = plt.axhline(y=-1, color='purple', linestyle='--')
```

```python
torch.norm(Kcol_2)
```

```python
torch.norm(Kcol_3)
```

<p>We've now determined that, in addition to being orthogonal, the columns of K have unit norm, therefore they are orthonormal.

To ensure that K is an orthogonal matrix, we would need to show that not only does it have orthonormal columns but it has orthonormal rows are as well. Since $$K^T$$!=K, we can't prove this quite as straightforwardly as we did with $$I_3$$.

One approach would be to repeat the steps we used to determine that K has orthogonal columns with all of the matrix's rows (please feel free to do so). Alternatively, we can use an orthogonal matrix-specific equation from the slides, $$A^TA=I$$, to demonstrate that K is orthogonal in a single line of code:  </p>

```python
torch.matmul(K.T, K)
```

<p>Notwithstanding rounding errors that we can safely ignore, this confirms that $$K^TK=I$$ and therefore K is an orthogonal matrix.</p>

```python
import matplotlib.pyplot as plt
```

```python
def plot_vectors(vectors, colors):
    """
    Plot one or more vectors in a 2D plane, specifying a color for each. 

    Arguments
    ---------
    vectors: list of lists or of arrays
        Coordinates of the vectors to plot. For example, [[1, 3], [2, 2]] 
        contains two vectors to plot, [1, 3] and [2, 2].
    colors: list
        Colors of the vectors. For instance: ['red', 'blue'] will display the
        first vector in red and the second in blue.
        
    Example
    -------
    plot_vectors([[1, 3], [2, 2]], ['red', 'blue'])
    plt.xlim(-1, 4)
    plt.ylim(-1, 4)
    """
    plt.figure()
    plt.axvline(x=0, color='lightgray')
    plt.axhline(y=0, color='lightgray')

    for i in range(len(vectors)):
        x = np.concatenate([[0,0],vectors[i]])
        plt.quiver([x[0]], [x[1]], [x[2]], [x[3]],
                   angles='xy', scale_units='xy', scale=1, color=colors[i],)
```

<h3>Eigendecomposition</h3>
<p>The <b>eigendecomposition</b> of some matrix <b>A</b> is $$A = VV^-1$$</p>
<p>Where:</p>
<ul>
    <li>As in examples above, <b>V</b> is the concatenation of all the eigenvectors of <b>A</b></li>
    <li> (upper-case ) is the diagonal matrix diag(). Note that the convention is to arrange the lambda values in descending order; as a result, the first eigenvalue (and its associated eigenvector) may be a primary characteristic of the matrix <b>A</b>.</li>
</ul>

```python
# This was used earlier as a matrix X; it has nice clean integer eigenvalues...
A = np.array([[4, 2], [-5, -3]]) 
A
```

```python
lambdas, V = np.linalg.eig(A)
```

```python
V
```

```python
Vinv = np.linalg.inv(V)
Vinv
```

```python
Lambda = np.diag(lambdas)
Lambda
```

<p>Confirm that $$A=VV^{-1}:$$</p>

```python
np.dot(V, np.dot(Lambda, Vinv))
```

<p>Eigendecomposition is not possible with all matrices. And in some cases where it is possible, the eigendecomposition involves complex numbers instead of straightforward real numbers.

<h3>Singular Value Decomposition (SVD)</h3>
<p>As on slides, SVD of matrix <b>A</b> is: $$A=UDV^T$$
Where:</p>
<ul>
    <li><b>U</b> is an orthogonal m x m matrix; its columns are the left-singular vectors of <b>A</b>.</li>
    <li><b>V</b> is an orthogonal n x n matrix; its columns are the right-singular vectors of <b>A</b></li>
    <li><b>D</b> is a diagonal m x n matrix; elements along its diagonal are the singular values of <b>A</b>.</li>
</ul>

```python
A = np.array([[-1, 2], [3, -2], [5, 7]])
A
```

```python
U, d, VT = np.linalg.svd(A) # V is already transposed
```

```python
U
```

```python
VT
```

```python
d
```

```python
np.diag(d)
```

<p><b>D</b> must have the same dimensions as <b>A</b> for $$UDV^T$$ matrix multiplication to be possible:</p>

```python
D = np.concatenate((np.diag(d), [[0, 0]]), axis=0)
D
```

```python
np.dot(U, np.dot(D, VT))
```

<p>SVD and eigendecomposition are closely related to each other:</p>
<ul>
    <li>Left-singular vectors of A = eigenvectors of $$AA^T$$.</li>
    <li>Right-singular vectors of A = eigenvectors of $$A^TA$$</li>
    <li>Non-zero singular values of A = square roots of eigenvalues of $$AA^T$$ = square roots of eigenvalues of $$A^TA$$</li>
</ul>

<h3>Segment 2: Computing Derivatives with Differentiation</h3>
<p>Let's bring back our ol' buddy $$y=x^2+2x+2:$$</p>

```python
def f(my_x):
    my_y = my_x**2 + 2 * my_x + 2
    return my_y
```

```python
y = f(x)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
_ = ax.plot(x,y)
```

<p>Let's identify the slope where, say, x=2
    
First let's determine what y is:</p>

```python
f(2)
```

<p>Cool. Let's call this point P, which is located at (2, 10):</p>

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(2, 10) # new
_ = ax.plot(x,y)
```

<p>The delta method uses the difference between two points to calculate slope. To illustrate this, let's define another point, Q where, say, x=5.</p>

```python
f(5)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(2, 10)
plt.scatter(5, 37, c = 'orange', zorder=3) # new
_ = ax.plot(x,y)
```

<p>To find the slope m between points P and Q:

$$m=(change in y)/(change in x)=(y)/(x)=(y_2-y_1)/(x_2-x_1)=(37-10)/(5-2)=27/3=9$$</p>

```python
m = (37-10)/(5-2)
m
```

<p>To plot the line that passes through P and Q, we can rearrange the equation of a line y = mx + b to solve for b: $$b = y - mx$$</p>

```python
b = 37-m*5
b
```

```python
line_y = m*x + b
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(2, 10)
plt.scatter(5, 37, c='orange', zorder=3)
plt.ylim(-5, 150) # new
plt.plot(x, line_y, c='orange') # new
_ = ax.plot(x,y)
```

<p>The closer Q becomes to P, the closer the slope m comes to being the true tangent of the point P. Let's demonstrate this with another point Q at x=2.1
Previously, our x between Q and P was equal to 3. Now it is much smaller: $$x = x_2 - x_1 = 2.1 - 2 = 0.1$$</p>

```python
f(2.1)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(2, 10)
plt.scatter(2.1, 10.61, c = 'orange', zorder=3)
_ = ax.plot(x,y)
```

```python
m = (10.61-10)/(2.1-2)
m
```

```python
b = 10.61 - m * 2.1
b
```

```python
line_y = m * x + b
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(2, 10)
plt.scatter(2.1, 10.61, c='orange', zorder=3)
plt.ylim(-5, 150)
plt.plot(x, line_y, c='orange', zorder=3)
_ = ax.plot(x,y)
```

<p>The closer Q becomes to P (i.e, x approaches 0), the clearer it becomes that the slope m at point P = (2, 10) is equal to 6.

Let's make x extremely small, 0.000001, to illustrate this:</p>

```python
delta_x = 0.000001
delta_x
```

```python
x1 = 2
y1 = 10
```

<p>Rearranging x = x2 - x1, we can calculate x2 for our point Q, which is now extremely close to P: $$x_2 = x_1 + x$$</p>

```python
x2 = x1 + delta_x
x2
```

<p>y2 for our point Q can be obtained with the usual function f(x): $$y_2 = f(x_2)$$</p>

```python
y2 = f(x2)
y2
```

<p>To find the slope m, we continue to use: $$m = y/x = (y_2-y_1)/(x_2-x_1)$$</p>

```python
m = (y2-y1) / (x2-x1)
m
```

<p>Boom! Using the delta method, we've shown that at point P, the slope of the curve is 6.

<p>Now, let's observe the differentiation equation in action:</p>

```python
def diff_demo(my_f, my_x, my_delta):
    return (my_f(my_x + my_delta) - my_f(my_x)) / my_delta
```

```python
deltas = [1, 0.1, 0.01, 0.001, 0.0001, 0.00001, 0.000001]
```

```python
for delta in deltas:
    print(diff_demo(f, 2, delta))
```

```python
for delta in deltas:
    print(diff_demo(f, -1, delta))
```
